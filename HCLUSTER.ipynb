```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Mall_customers.csv')
```

```python
data
```

```python
x = data.iloc[:,[3,4]].values
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))
```

```python
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(x)
```

```python
plt.scatter(x[y_hc == 0,0], x[y_hc == 0,1], s= 100, label = 'Cluster1', c = 'Red')
plt.scatter(x[y_hc == 1,0], x[y_hc == 1,1], s= 100, label = 'Cluster2', c = 'Green')
plt.scatter(x[y_hc == 2,0], x[y_hc == 2,1], s= 100, label = 'Cluster3', c = 'purple')
plt.scatter(x[y_hc == 3,0], x[y_hc == 3,1], s= 100, label = 'Cluster4', c = 'Blue')
plt.scatter(x[y_hc == 4,0], x[y_hc == 4,1], s= 100, label = 'Cluster5', c = 'Orange')
```

```python
plt.title('Clusters')
plt.xlabel('Income')
plt.ylabel('spending')
plt.legend()
plt.show()
```
